Column Transformation

```python
data['Attrition']=data['Attrition'].replace({"Yes":1,"No":0})
```

```python
data
```

Overall Attrition Rate

```python
total_employees = len(data)
attrition_count = data["Attrition"].sum()
```

```python
attrition_count
```

```python
attrition_rate = attrition_count / total_employees
print("Overall Attrition Rate: {:.2%}".format(attrition_rate))
```

Attrtion By Department

```python
department_attrition = data.groupby("Department")["Attrition"].sum().reset_index()
print(department_attrition)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(department_attrition["Department"], department_attrition["Attrition"], color='skyblue')
plt.title('Total Attritions by Department')
plt.xlabel('Department')
plt.ylabel('Total Attritions')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Attrition By Marital Status and Gender

```python
marital_gender_attrition=data.groupby(["MaritalStatus","Gender"])["Attrition"].sum().reset_index()
print(marital_gender_attrition)
```

```python
pivot_marital_gender_attrition = data.pivot_table(index=["MaritalStatus", "Gender"],
                                                 values="Attrition", aggfunc="sum").reset_index()
plt.figure(figsize=(10, 6))
plt.bar(pivot_marital_gender_attrition["MaritalStatus"] + " - " + pivot_marital_gender_attrition["Gender"],
        pivot_marital_gender_attrition["Attrition"], color='skyblue')
plt.title('Total Attritions by MaritalStatus and Gender')
plt.xlabel('MaritalStatus - Gender')
plt.ylabel('Total Attritions')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Average Age of Employees who are departing rom company

```python
departing_employees = data[data["Attrition"] == 1]
average_age_departing = departing_employees["Age"].mean()
print("Average Age of Departing Employees: {:.2f}".format(average_age_departing))
```

Average Tenure of Employees who are departing

```python
average_tenure_departing = departing_employees["TotalWorkingYears"].mean()
print("Average Tenure of Departing Employees: {:.2f} years".format(average_tenure_departing))
```

Educational Level of employees who are departing

```python
education_attrition = data.groupby("Education")["Attrition"].sum().reset_index()
print(education_attrition)
```

Overtime Analysis

```python
attrition_with_overtime = data[data["OverTime"] == "Yes"]["Attrition"].mean()
attrition_without_overtime = data[data["OverTime"] == "No"]["Attrition"].mean()

print("Attrition Rate for Employees Working Overtime: {:.2%}".format(attrition_with_overtime))
print("Attrition Rate for Employees Not Working Overtime: {:.2%}".format(attrition_without_overtime))
```

Job Satisfaction

```python
job_satisfaction_attrition = data.groupby("JobSatisfaction")["Attrition"].sum().reset_index()
print(job_satisfaction_attrition)
```

Business Travel

```python
business_travel_attrition = data.groupby("BusinessTravel")["Attrition"].sum().reset_index()
print(business_travel_attrition)
```

CompaniesWorked

WorkLife Balanace

```python
work_life_balance_attrition = data.groupby("WorkLifeBalance")["Attrition"].sum().reset_index()
print(work_life_balance_attrition)
```

```python
age_by_overtime = data.groupby("OverTime")["Age"].mean().reset_index()
print(age_by_overtime)
```

```python
plt.figure(figsize=(10, 6))

# Scatter points for employees who work overtime (Overtime: Yes)
plt.scatter(data[data["OverTime"] == "Yes"]["Age"], data[data["OverTime"] == "Yes"].index,
            color='red', label='Overtime: Yes', alpha=0.5)

# Scatter points for employees who do not work overtime (Overtime: No)
plt.scatter(data[data["OverTime"] == "No"]["Age"], data[data["OverTime"] == "No"].index,
            color='blue', label='Overtime: No', alpha=0.5)

plt.title('Relation between Overtime and Age')
plt.xlabel('Age')
plt.ylabel('Employee Index')
plt.legend()
plt.show()
```

```python
distance_by_attrition = data.groupby("Attrition")["DistanceFromHome"].mean().reset_index()
print(distance_by_attrition)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(distance_by_attrition["Attrition"], distance_by_attrition["DistanceFromHome"], color=['blue', 'red'])
plt.title('Average Distance from Home by Attrition Status')
plt.xlabel('Attrition Status')
plt.ylabel('Average Distance from Home (Miles)')
plt.show()
```

Attrition by Monthly Income

```python
attrition_yes = data[data["Attrition"] == 1]
attrition_no = data[data["Attrition"] == 0]
# Create a histogram to compare the distribution of monthly incomes
plt.figure(figsize=(10, 6))
plt.hist(attrition_yes["MonthlyIncome"], bins=20, alpha=0.5, color='red', label='Attrition: Yes')
plt.hist(attrition_no["MonthlyIncome"], bins=20, alpha=0.5, color='blue', label='Attrition: No')
plt.title('Attrition by Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
